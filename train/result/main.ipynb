```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline, BSpline
```

```python
def plotLoss():
    Loss = []
    with open('loss.text', 'r') as f:
        Loss_str = f.read()
        Loss = Loss_str.split(" ")
    Loss = np.array(Loss[29:-2]).astype(float)

    # get avg loss
    Loss_avg = []
    for i in range (0, len(Loss) - 50, 50):
        Loss_avg.append(np.mean(Loss[i:i + 50]))

    # plot figure
    x = range(0, len(Loss_avg))
    xnew = np.linspace(min(x), max(x), 1000) 

    spl = make_interp_spline(x, Loss_avg, k = 3)  # type: BSpline
    power_smooth = spl(xnew)

    plt.figure(figsize=(10, 5), dpi = 100)
    plt.xlabel("Episode")
    plt.ylabel("Avg loss")
    plt.plot(xnew, power_smooth)
    plt.savefig('loss.png', bbox_inches = 'tight')
    plt.show()
```
